```python
students = []
def assign_grade(avg):
    if avg>=75:
        return "Distinction"
    elif avg>=60:
        return "First Class"
    elif avg>=40:
        return "Second Class"
    else:
        return "Fail"
        
def add_student():
    roll=input("Enter Roll Number: ")
    name=input("Enter Name: ")
    branch=input("Enter Branch: ")
    m1=float(input("Enter Subject 1 Marks: "))
    m2=float(input("Enter Subject 2 Marks: "))
    m3=float(input("Enter Subject 3 Marks: "))
    avg=(m1+m2+m3)/3
    grade=assign_grade(avg)
    student = {
        "roll": roll,
        "name": name,
        "branch": branch,
        "marks": [m1, m2, m3],
        "average": avg,
        "grade": grade
    }
    students.append(student)
    print("Student Added Successfully!")

def view_students():
    if not students:
        print("No records found.")
        return

    for s in students:
        print("\nRoll No:", s["roll"])
        print("Name:", s["name"])
        print("Branch:", s["branch"])
        print("Marks:", s["marks"])
        print("Average:", round(s["average"], 2))
        print("Grade:", s["grade"])

def search_student():
    key=input("Enter Roll Number or Name: ")

    found=False

    for s in students:
        if s["roll"] == key or s["name"].lower() == key.lower():
            print(s)
            found = True

    if not found:
        print("Student Not Found")

def update_student():
    roll=input("Enter Roll Number to Update: ")

    for s in students:
        if s["roll"]==roll:
            s["name"]= nput("Enter New Name: ")
            s["branch"]=input("Enter New Branch: ")

            m1=float(input("Enter Subject 1 Marks: "))
            m2=float(input("Enter Subject 2 Marks: "))
            m3=float(input("Enter Subject 3 Marks: "))
            s["marks"]=[m1, m2, m3]
            s["average"]=(m1 + m2 + m3) / 3
            s["grade"]=assign_grade(s["average"])
            print("Record Updated Successfully!")
            return

    print("Student Not Found")

def delete_student():
    roll=input("Enter Roll Number to Delete: ")

    for s in students:
        if s["roll"]==roll:
            students.remove(s)
            print("Record Deleted Successfully!")
            return

    print("Student Not Found")

def class_average():
    if not students:
        print("No records available.")
        return

    total=sum(s["average"] for s in students)
    avg=total/len(students)

    print("Class Average =", round(avg, 2))

def assign_grades_all():
    for s in students:
        s["grade"]=assign_grade(s["average"])

    print("Grades Assigned Successfully!")

def filter_students():
    grade=input("Enter Grade Category: ")
    found=False

    for s in students:
        if s["grade"].lower()==grade.lower():
            print(s)
            found=True

    if not found:
        print("No Students Found")

def find_topper():
    if not students:
        print("No records available.")
        return

    highest=max(s["average"] for s in students)

    print("Class Topper(s):")

    for s in students:
        if s["average"]==highest:
            print(s)

while True:
    print("\n===== Student Performance Management System =====")
    print("1. Add New Student")
    print("2. View All Students")
    print("3. Search Student")
    print("4. Update Student Record")
    print("5. Delete Student Record")
    print("6. Calculate Class Average")
    print("7. Assign Grades to All Students")
    print("8. Filter Students by Grade")
    print("9. Find Class Topper")
    print("10. Exit")
    choice=int(input("Enter your choice: "))
    if choice==1:
        add_student()
    elif choice==2:
        view_students()
    elif choice==3:
        search_student()
    elif choice==4:
        update_student()
    elif choice==5:
        delete_student()
    elif choice==6:
        class_average()
    elif choice==7:
        assign_grades_all()
    elif choice==8:
        filter_students()
    elif choice==9:
        find_topper()
    elif choice==10:
        print("Exiting Program...")
        break
    else:
        print("Invalid Choice")
```
